# Q3

```python
df_seoul_dermatology = df[(df["상권업종소분류명"].str.contains("피부|성형")) & (df["시도명"] == "서울특별시")]
df_seoul_dermatology["시군구명"].value_counts()
```

```python
df_seoul_hospital = df[(df["시도명"] == "서울특별시") & (df["상권업종중분류명"] == "병원")].copy()
df_seoul_hospital["시군구명"].value_counts()
```

```python
ratio = df_seoul_dermatology["시군구명"].value_counts() / df_seoul_hospital["시군구명"].value_counts()
round(ratio.sort_values(ascending=False), 2)
```

# Q4

```python
import folium

df_geriatric_hospital = df[df["상권업종소분류명"] == "노인/치매병원"]
df_geriatric_hospital
```

```python
map = folium.Map(location=[df_geriatric_hospital["위도"].mean(), df_geriatric_hospital["경도"].mean()], zoom_start=7)
```

```python
for n in df_geriatric_hospital.index:
    name = df_geriatric_hospital.loc[n, "상호명"]
    address = df_geriatric_hospital.loc[n, "도로명주소"]
    popup = f"{name}-{address}"
    location = [df_geriatric_hospital.loc[n, "위도"], df_geriatric_hospital.loc[n, "경도"]]
    folium.Marker(
        location = location,
        popup = popup,
    ).add_to(map)
map
```
