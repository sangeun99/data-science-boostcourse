### 여러 파일을 한번에 로드하는 방법

```python
file_csv = glob("../data/store/*_*_*.csv")
file_csv
```

```python
file_list = []
for file_csv_name in file_csv:
  # print(file_csv_name)
  temp = pd.read_csv(file_csv_name, low_memory=False)
  file_list.append(temp)
```

```python
file_list
```

```python
len(file_list)
```

```python
df = pd.concat(file_list)
df.shape
```
